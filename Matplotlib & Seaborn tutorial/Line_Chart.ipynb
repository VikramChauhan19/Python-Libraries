# Data Visualization:
It is the graphic representation of data. It involves producing images that communicate relationships among the represented data to viewers.We'll use Python libraries Matplotlib and Seaborn to learn and apply some popular data visualization techniques

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Line Chart:
The line chart is one of the simplest and most widely used data visualization techniques. A line chart displays information as a series of data points or markers connected by straight lines

```python
yield_apple = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931] #yield of apple over 6 years
plt.plot(yield_apple); #use for line chart, ;(semicolon) is for removing unintended output
```

### Customizing the X-axis

```python
years = [2010, 2011, 2012, 2013, 2014, 2015]
plt.plot(years, yield_apple) #(X-axis, Y-axis)
```

### Axis Labels
We can add labels to the axes to show what each axis represents using the plt.xlabel and plt.ylabel methods.

```python
plt.plot(years, yield_apple) #any plt fuction in same cell apply to same plot
plt.xlabel('years')
plt.ylabel('yield_apple');
```

### Plotting Multiple Lines
You can invoke the plt.plot function once for each line to plot multiple lines in the same graph. Let's compare the yields of apples vs. oranges


```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
plt.xlabel('Yield')
plt.plot(years, apples)
plt.plot(years, oranges)
plt.ylabel("YIeld(tons per hectare)")
```

### Chart Title and Legend
To differentiate between multiple lines, we can include a legend within the graph using the plt.legend function. We can also set a title for the chart using the plt.title function

```python
plt.xlabel('years')
plt.plot(years, apples)
plt.plot(years, oranges)
plt.ylabel("YIeld(tons per hectare)")
plt.title("Crop Yield in Kanto") #add titlr for whole graph
plt.legend(["apples", "oranges"]); #give label to defferentiate between multiple lines
```

### Line Markers:
We can also show markers for the data points on each line using the marker argument of plt.plot. Matplotlib provides many different markers, like a circle, cross, square, diamond, etc. You can find the full list of marker types here: https://matplotlib.org/3.1.1/api/markers_api.html .

```python
plt.plot(years, apples, marker = 'o')
plt.plot(years, oranges, marker = 'v')
plt.xlabel('years')
plt.ylabel('Yield ')
plt.title("Crop Yeild in kanto")
plt.legend(['apples', 'oranges']);
```

### Styling Lines and markers
It provide various arguments for styling and markers:

1 color or c: Set the color of the line

2 linestyle or ls: Choose between a solid or dashed line

3 linewidth or lw: Set the width of a line

4 markersize or ms: Set the size of markers

5 markeredgecolor or mec: Set the edge color for markers

6 markeredgewidth or mew: Set the edge width for markers

7 markerfacecolor or mfc: Set the fill color for markers

8 alpha: Opacity of the plot

Check out the documentation for plt.plot to learn more: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot .




```python
plt.plot(years, apples, marker = 'o', c = 'b', ls = '-', lw = '2',ms = 8, mew = 2, mec = 'navy')
plt.plot(years, oranges, marker = 'v', c = 'r', ls = '--', lw = '3', ms = 10, alpha = 0.5 )
plt.xlabel('years')
plt.ylabel('Yield ')
plt.title("Crop Yeild in kanto")
plt.legend(['apples', 'oranges']);
```

The fmt argument provides a shorthand for specifying the marker shape, line style, and line color.

fmt = '[marker][line][color]'

```python
plt.plot(years, apples, 's-b')
plt.plot(years, oranges, 'o--r')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']);
```

```python
plt.plot(years, oranges, 'or') #markers will not be drawn
plt.title("Yield of Oranges (tons per hectare)");
```

### Changing the Figure Size


```python
plt.figure(figsize = (12, 6)) #here we are giving tuple
plt.plot(years, oranges, 'or')
plt.title("Yield of Oranges (tons per hectare)");
```
